```python
import pandas as pd
import matplotlib.pyplot as plt
people = ['Mani','Surya','Mukesh','Meghanadh']
age = [25,30,35,40]
height =[145,151,165,173]
weight =[45,55,65,75]
data = {'Name':people,'Age':age,'height':height,'Weight':weight}
df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(10,6))
plt.bar(df['Name'],df['height'],color='green')
plt.title("Bar chart of individual\'Height")
plt.xlabel("Name")
plt.ylabel("Height(cm)")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
